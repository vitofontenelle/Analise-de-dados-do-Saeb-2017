```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
from scipy.stats import kurtosis
from scipy.stats import skew
```

```python
df = pd.read_csv('amostra_221014489.csv', encoding='utf-8')
```

```python
df['COMPUTADOR'].unique()
```

```python
# Verificando os valores únicos para encontra aqueles que não deram repostas.
df['RACA_COR'].unique()
```

```python
# Gráfico 1 de pizza para variável RACA_COR
raca_qtd = df.RACA_COR.replace(' ', 'Não quero declarar').value_counts() # Aqui passei os valores vazios(os que não respoderam) para 'Não quero declarar'
cores = ['#F8F4E3', '#CE8964',  '#63595C',  '#297045', '#A1B5D8', '#F5CB5C']
fig, ax = plt.subplots(figsize=(9,6))
ax.pie(raca_qtd, labels=raca_qtd.index, autopct='%1.1f%%', colors=cores)
plt.title('Raca/cor autodeclarada dos alunos')
```

```python
# Tabela 1 para a variável RACA_COR
tabela_raca = pd.DataFrame({
    'Quantidade': raca_qtd.values,
    'Percentual(%)': ((raca_qtd / raca_qtd.sum())*100).round(1)
})

print(tabulate(tabela_raca, headers='keys', tablefmt='pretty'))
```

```python
# Gráfico 2 por região
regiao_qtd = df['REGIAO'].value_counts()
fig, ax = plt.subplots(figsize=(9,6))
ax.pie(regiao_qtd, labels=regiao_qtd.index, autopct='%1.1f%%', colors=cores)
plt.title('Região onde as provas foram realizadas')
```

```python
# Tabela 2 para a variável REGIAO
tabela_regiao = pd.DataFrame({
    'Quantidade': regiao_qtd.values,
    'Percentual(%)': ((regiao_qtd / regiao_qtd.sum())*100).round(1)
})

print(tabulate(tabela_regiao, headers='keys', tablefmt='pretty'))
```

```python
# Gráfico 3 para a variável LOCALIZACAO
localizacao_qtd = df['LOCALIZACAO'].value_counts(normalize=True)
sns.barplot(localizacao_qtd, palette='pastel')
plt.ylabel('Frequência')
plt.xlabel('Localização')
plt.title('Localização da escola do estudante')
```

```python
# Tabela 3 para a variável LOCALIZACAO
loc_qtd = df['LOCALIZACAO'].value_counts()
tabela_localizacao = pd.DataFrame({
    'Quantidade': loc_qtd.values,
    'Percentual(%)': ((loc_qtd / loc_qtd.sum())*100).round(1)
})

print(tabulate(tabela_localizacao, headers='keys', tablefmt='pretty'))
```

```python
# Grafico 4 para a variável DEPENDENCIA_ADM
dependencia_qtd = df['DEPENDENCIA_ADM'].value_counts()
sns.barplot(dependencia_qtd, palette='pastel')
plt.ylabel('Frequência')
plt.xlabel('Dependencia')
plt.title('Dependencia da escola dos alunos')
```

```python
# Tabela 4 para a variável DEPENDENCIA_ADM
tabela_dependencia = pd.DataFrame({
    'Quantidade': dependencia_qtd.values,
    'Percentual(%)': ((dependencia_qtd / dependencia_qtd.sum())*100).round(1)
})

print(tabulate(tabela_dependencia, headers='keys', tablefmt='pretty'))
```

```python
# Gráfico 5 para a variável COMPUTADOR
pc_qtd = df['COMPUTADOR'].replace(' ', 'Não respondeu').value_counts()
sns.barplot(pc_qtd, palette='pastel')
plt.xticks(rotation=45)
plt.xlabel('Computador')
plt.ylabel('Frequência')
plt.title('Quantidade de computador na casa do aluno')
```

```python
# Tabela 5 para a variável COMPUTADOR
tabela_computador = pd.DataFrame({
    'Quantidade': pc_qtd.values,
    'Percentual(%)': ((pc_qtd / pc_qtd.sum())*100).round(1)
})

print(tabulate(tabela_computador, headers='keys', tablefmt='pretty'))
```

```python
# Estatísticas das notas de matemática
curtose = kurtosis(df['NOTA_MT'], fisher=False) # Usando Pearson
mediana = np.median(df['NOTA_MT'])
maximo = df['NOTA_MT'].max()
minimo = df['NOTA_MT'].min()
media = np.mean(df['NOTA_MT'])
assimetria = skew(df['NOTA_MT'])
print(f'curtose = {curtose:.2f}\nmediana = {mediana:.2f}\nmáximo = {maximo:.2f}\nmínimo = {minimo:.2f}\nmédia = {media:.2f}\nassimetria = {assimetria:.3f}')
```

```python
# Estatísticas das notas de português
curtose = kurtosis(df['NOTA_LP'], fisher=False) # Usando Pearson
mediana = np.median(df['NOTA_LP'])
maximo = df['NOTA_LP'].max()
minimo = df['NOTA_LP'].min()
media = np.mean(df['NOTA_LP'])
assimetria = skew(df['NOTA_LP'])
print(f'curtose = {curtose:.2f}\nmediana = {mediana:.2f}\nmáximo = {maximo:.2f}\nmínimo = {minimo:.2f}\nmédia = {media:.2f}\nassimetria = {assimetria:.3f}')
```
